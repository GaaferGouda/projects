```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation constants
target_distance = 500  # Target distance in meters
thrust_force = 50      # Thrust force (N)
mass = 5               # Rocket mass (kg)
dt = 0.1               # Time step (s)

# Variables
position = 0           # Initial position
velocity = 0           # Initial velocity
time = 0               # Time
positions = []         # To store positions for plotting

def simulate_rocket(burn_time):
    global position, velocity, time
    position, velocity, time = 0, 0, 0
    positions.clear()

    while position < target_distance * 1.5:  # Continue past the target to show overshooting
        if time < burn_time:
            acceleration = thrust_force / mass
        else:
            acceleration = 0

        velocity += acceleration * dt
        position += velocity * dt
        time += dt

        positions.append(position)

    return position

# Run simulation
burn_time_values = [5, 7, 9]  # Different burn times to demonstrate overshooting
plt.figure(figsize=(8, 5))

for burn_time in burn_time_values:
    simulate_rocket(burn_time)
    plt.plot(np.arange(0, len(positions) * dt, dt), positions, label=f'Burn time {burn_time} seconds')

# Mark target distance
plt.axhline(y=target_distance, color='r', linestyle='--', label='Target')
plt.title('Rocket Overshoot Simulation')
plt.xlabel('Time (seconds)')
plt.ylabel('Position (meters)')
plt.legend()
plt.grid(True)
plt.show()
```
