## More Visualizations with Matplotlib ðŸŽ¨

You can make more complex visuals using matplotlib using other methods and by utlizing other libraries in tandem such as `math` and `numpy`

````python
import math
import numpy as np
````

The following code snippet visualizes the `sin` function using the `math` library.
NOTE: we use the `linspace` method using numpy to create evenly spaced values between 0 and 2 pi.

````python
x = np.linspace(0, 2*math.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x)')
plt.grid(True)
plt.show()
````

````python
# Define data for planets
planet_names = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
semi_major_axis = [0.39, 0.72, 1.0, 1.52, 5.2, 9.58, 19.22, 30.05]  # in AU (astronomical units)
orbit_period = [0.24, 0.62, 1.0, 1.88, 11.86, 29.46, 84.01, 164.8]  # in years
````

# Plotting Orbits in 2D
Since the eccentricity and inclination of each of the planets (see above) is quite low, we can treat each of their orbital paths as circles.

The code below demonstrates how to visualize each of the planet orbital paths using `matplotlib` and `numpy`.

We place the Sun at the center using the method:

```
ax.scatter(0, 0, color='yellow', s=100, label='Sun')
```

We plot each of the planet orbits by treating them as circles with two core components:
- Orbital Radius (aka Semi Major Axis)

```
orbit_radius = semi_major_axis[i]
```

- Orbital Angle:
```
# generates 100 equally spaced points between 0 and 2Ï€, covering a full circle 
orbit_theta = np.linspace(0, 2*np.pi, 100)

# calculates the x-coordinates of points on the orbit
x = orbit_radius * np.cos(orbit_theta)

# calculates the y-coordinates of points on the orbit
y = orbit_radius * np.sin(orbit_theta)
```

````python
plt.style.use('dark_background')
# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Plot the Sun at the center
ax.scatter(0, 0, color='yellow', s=100, label='Sun')

# Plot orbits of planets
for i, planet in enumerate(planet_names):
    orbit_radius = semi_major_axis[i]
    orbit_theta = np.linspace(0, 2*np.pi, 100)
    x = orbit_radius * np.cos(orbit_theta)
    y = orbit_radius * np.sin(orbit_theta)
    ax.plot(x, y, label=planet)

# Set aspect ratio to equal, and add labels
ax.set_aspect('equal', 'box')
ax.set_xlabel('Distance (AU)')
ax.set_ylabel('Distance (AU)')
ax.set_title('Solar System')
ax.legend()

plt.grid(True)
plt.show()
````
